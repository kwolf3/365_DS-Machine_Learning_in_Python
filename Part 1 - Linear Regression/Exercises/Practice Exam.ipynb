```python
import pathlib
import pandas as pd
import numpy
import statsmodels.api as sm
```

```python
data_folder = pathlib.Path.cwd().parent/"Data"
data = pd.read_csv(data_folder/"Startups.csv")
data
```

```python
data.describe(include = "all")
```

```python
y = data["Profit"]
x1 = data["Marketing Expenditure"]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

```python
new_data = pd.DataFrame({'const': 1, 'Marketing Expenditure': 325000}, index = [0])
new_data
```

```python
pred = results.predict(new_data)
pred
```
